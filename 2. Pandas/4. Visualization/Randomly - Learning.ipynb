# → New arrived, `df.asof()` 

This function will return the LAST non NaN row.

    Example: I have 1, 2, 3, 4, 5 rows and 2 and 4 have NaN (forget about the columns for a sec - there is just a single col)
             then if I write df.asof([1,2,3]) then it will return:
                                      ↑ ↑ ↑
             Row 1 data for 1 --------| | |
             Row 1 data for 2 ----------| |
             Row 3 data for 3 ------------|
             
             Why not for 2? Because row 2 is having NaN. So it will find the row above that 2 row in this case, and will return.
             
             • What if No row found?
             Then the NaN will be returned. 
             
             • What if my first row is having NaN? Will it still check above?
             No, as none of the rows are there above the first row, so NaN will be returned.
             
             • What if I put the row index out of the available i.e, 1000 or 4000?
             Then it will search from the last row, and will try to find the first row from the last, which is non NaN.

For a bit better understanding look below ↓

````python
df = pd.DataFrame(np.random.randint(0, 100, (5, 5)))
df
````

````python
df.iloc[[0,3,4], [1,2,4]] = np.nan
````

````python
df
````

````python
df.asof([0,4,3])
````

See? For row 4 and 3 - it has returned the data of row 2 which has non null.

And if you want to go for a perticular column, use `subset= ['']` so it will check for the NaN in that/those column(s) only.

____
#### And yes, while going for `.asof` it checks the provided list on the INDEX and the nulls (NaN) in the data.

Here we will get a chance to explore the use of `asof` and also working with the date, data type.

This example requires the data from the stockmarket - let's get it.

````python
import pandas_datareader
````

````python
IDEA = pandas_datareader.DataReader('IDEA', data_source= 'yahoo')
IDEA
````

````python
from datetime import datetime
````

Now, from the data, I want to plot the data all about the opening of IDEA daily... So, I would only store single column.

````python
IDEA_opens = IDEA['Open']
````

````python
IDEA_opens.plot()
````

Now, the annotation.  
_  <br>
I want to plot on some dates, I am going to make the list of those dates and labels. I will include...
* My birthday
* The day I joined market
* The day I bought the shares

````python
dates = \
[ 
  (datetime(2021, 4, 25), 'My Bitrthday'),
  (datetime(2021, 3, 28), 'I joined here'),
  (datetime(2021, 4, 9), 'I bought IDEA')
]
````

Now, need to understand a little more.<br>
_ <br>
Now, the reason why we need to use asof, it will return the value, on that date - or if not date found, the just above will be returned. (Here, it will be using the third question discussed in asof)

````python
dates[0][0]
````

````python
# Just checking...
IDEA_opens['2021-4-25']
````

### SEE? On my birthday, the stockmarket was not open ie. data is not available.
# BUT! See this ↓

````python
IDEA_opens.asof(dates[0][0])
````

# On my birthdat ↑ it is showing some data.
That means it should be showing data from 24 or 23 date.

````python
IDEA_opens['2021-4-23']
````

Yeah ↑ it is from 23rd April.

# THIS
<font style= "font-size:80px"> IS </font> <br>
<font style= "font-size:100px"> HOW </font> <br>
<font style= "font-size:150px"> asof is useful </font>

````python
for date, label in dates:
    print(f'Date: {date} - Label: {label}')
````

````python
ax = plt.axes()
IDEA_opens.plot(ax= ax)
for date, label in dates:
    ax.annotate(label, 
                xy= (date, IDEA_opens.asof(date)))
````

It does, but! It is 4 years' data - so we need to FOCUS!

````python
IDEA_opens = IDEA_opens['2021-1' : '2021-6']
````

````python
ax = plt.axes()
IDEA_opens.plot(ax= ax)
for date, label in dates:
    ax.annotate(label, 
                xy= (date, IDEA_opens.asof(date)))
````

Let's be a bit clearer...

````python
ax = plt.axes()
IDEA_opens.plot(ax= ax)
for date, label in dates:
    ax.annotate(label, 
                xy= (date, IDEA_opens.asof(date)),
                xytext= (date, IDEA_opens.asof(date) + 0.04),
                arrowprops= dict(facecolor= "green", headwidth= 5, width= 2),
                horizontalalignment= 'left',
                verticalalignment= 'top',
                rotation= 45)
````
