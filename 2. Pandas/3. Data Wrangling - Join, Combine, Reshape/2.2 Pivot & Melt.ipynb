```python
def stylize(df, rows, cols):
    df_style = DataFrame('', index= df.index, columns= df.columns)
    df_style.iloc[rows, cols] = "background: #49f2ef"
    return df_style
```

### Now, let's check in online discussions and try to demystify them here...

### Answer 1:
* pivot_table is a **generalization** of pivot that **can handle duplicate values** for one pivoted index/column pair. Specifically, you can give pivot_table **a list of aggregation functions** using keyword argument aggfunc. The default aggfunc of pivot_table is **numpy.mean**. <font size=5> ✓ </font>

* pivot_table also **supports using multiple columns for the index** and column of the pivoted table. A hierarchical index will be automatically generated for you.

```python
# Solving the second point.
df.pivot(index= ['year', 'realgdp'], columns= ['quarter', 'realgovt'], values= 'pop')
```

So the second point is half truth. `MULTIINDEX` creation is SUPPORTED IN **BOTH TYPES**. ✓

# 

### ANSWER 2:


* `pivot_table` will **only allow numerical types** as "values=", whereas `pivot` will **take string** types as "values=".

```python
data1 = ['Aayush', 'Aayush', 'Aayush', 'Shah', 'Shah', 'Aayush']
data2 = [2017, 2018, 2019, 2017, 2018, 2020]
height = np.random.randint(100, 200, 6)
```

```python
temp = DataFrame({'name': data1,
                 'year': data2,
                 'height': height})
temp
```

**NOTE:** In the following examples, I have intentionally taken the NAME (str dtype) as the VALUES parameter.

```python
# Using pivot
temp.pivot(index= 'year', columns= 'height', values= 'name')
```

Works.

```python
# Using pivot_table
temp.pivot_table(index= 'year', columns= 'height', values= 'name')
```

Yes. Didn't work.  
The reason is quite obvioius. The `pivot_table` ***aggregates*** the values. So it can not take a MEAN of str values.

**But see below ↓**

```python
# Using pivot_table
temp.pivot_table(index= 'year', columns= 'height', values= 'name', aggfunc= 'max')
```

## Still using `pivot_table` we got the result. Why? 
It is because, we have changed the `aggfunc`. Now there is not any kind of mathematical operation, so we can have the data like this here.
<center> . . . </center>
    
Now, the statement:  
* pivot_table will only allow numerical types as "values=", whereas pivot will take string types as "values=". <font size= 4> X </font><br>
Because it is not entirely true.

### Some summary:
* Pivot is to get data as is from the table (no calculation)
* Pivot fails when duplicate index found
* To handle duplicates, we have pivot_table (yes calculation)
* If you can make pivot table of a table with `pivot`, then there is no need of `pivot_table` (both will return same)

# 

### Perspective:

1.<br>
**pivot** is a `set_index` operation followed by `unstack`.

2.<br>
**pivot_table** is `groupby` followed by some agg function and `unstack`.

# 

# Proving - 1
`pivot` == `set_index().unstack()`

```python
df.pivot('year', 'quarter', 'pop')[:5]
```

```python
df.set_index(['year', 'quarter'])['pop'].unstack()[:5]
```

# ***#PROVED***

# 

# Proving - 2
`pivot_table` == `groupby([index, columns])[values].agg().unstack()`

```python
df.pivot_table(index= 'year', columns= 'quarter', values= 'pop')[:5]
```

```python
df.groupby(['year', 'quarter'])['pop'].mean().unstack()[:5]
```

# ***#PROVED***

(Yes, in four of them are having the same result, as the data I have chosen has no duplicates - but I believe in the rule and that will work in other scenarios too)
