## → Merging on hierarchical index 

```python
index = pd.MultiIndex.from_product([['Bhuj', 'Ahmedabd', 'USA', 'London', 'World'], [2000, 2010, 2020]])
```

```python
df1 = pd.DataFrame(np.random.randint(1_00_000, 100_00_000, 15), index= index, columns= ['pop'])
df1
```

```python
df2 = pd.DataFrame({"area": np.random.choice(['ABCD', 'BCAD', 'AASS', 'BBCC'], 15),
                    "year": np.random.choice([2000, 2010, 2020], 15),
                   "city": np.random.choice(['Bhuj', 'Ahmedabad', 'USA', 'London', 'World'], 15)})

df2
```

```python
# now merging
pd.merge(df1, df2, left_index= True, right_on= ['city', 'year'])
```

Yes, of course - this will perform the many - to - many as the data is randomly choosen and repeating matches. But the thing that you need to notice is that - the merge has performed!

### What if I only want to merge only on the city name from the multi index?

```python
pd.merge(df1.reset_index(), df2, left_on= 'level_0', right_on= 'city')
```

Aha! Works!

## → No matter what the index is, I want to concatenate both of them

```python
df1 = pd.DataFrame(np.random.randint(0, 10, 5), index= [1,5,3,2,4])
df2 = pd.DataFrame(np.random.randint(0, 10, 5), index= [11,55,33,22,44]) 
```

```python
# See? This is I want to overcome
pd.concat([df1, df2], axis= 1)
```

```python
# If we try this                    ↓
pd.concat([df1, df2], axis= 1, ignore_index= True)
```

Simply doesn't work for the rows, but instead it has changed the COLUMN names (From 0, 0 to 0, 1) 

    `ignore_index` works for the concatenated axis. Here axis=1, so it has reset the index there (columns')
    I guess - we need to manually reset the index and then do the work.

```python
# So...
pd.concat([df1, df2.reset_index(drop= True)], axis= 1, ignore_index= True)
```

See? This works. Or you can use `df2.index = df1.index[:df2.shape[0]]`. Sounds silly but it can be a patch!

## → `combine_first` - An amazing tool 

* This is tool is like the np.where for pandas and is amazing while working with DF.  
* You can work with np.where in the series, but when it comes to the DF, np.where would fail.
* Then it comes to the more flexible tool specially mae for DF

```python
left = pd.DataFrame(np.random.randint(0, 100, (5,5)))
right = pd.DataFrame(np.random.randint(0, 100, (3,3)))
```

Try to `focus` here. The *left* df has 5x5 values and *right* df has 3x3 - but still the things would workout

Here, the main reason to use this combine_first is that, **Let's say** I want to fill out all the values in LEFT when there is NaN from the RIGHT df.

```python
# Making NaN values!
left.iloc[::2, ::2] = np.nan
```

```python
left
```

```python
right
```

```python
left.iloc[[0, 2, 4], [4]]
```

```python
# Now, using some magic!
def styling_specific_cell(df, rows, cols):
    df_styler = pd.DataFrame('', index= df.index, columns= df.columns)
    df_styler.iloc[rows, cols] = 'background-color: green'
    df_styler.iloc[[0, 2, 4], [4]] = 'background-color: red'
    df_styler.iloc[[4], [0, 2]] = 'background-color: red'
    return df_styler
    
left.style.apply(styling_specific_cell, rows= [0, 2], cols= [0, 2], axis= None)
```

So now, green cells will get their values from the RIGHT. The other cells with red color, won't get values from RIGHT as the RIGHT df itself doesn't have any kind of data available to fill in.

```python
left.combine_first(right).style.apply(styling_specific_cell, rows= [0, 2], cols= [0, 2], axis= None)
```

See? How cool!<br>
*(Cross checking)*
