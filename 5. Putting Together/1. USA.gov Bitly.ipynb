# 

### Windows users from Each timezones

```python
available_agents = df[df.a.notnull()].copy()
available_agents
```

```python
import re
```

```python
available_agents.a.str.contains('Windows', flags= re.IGNORECASE)
```

This is the **perfect** time to use `np.where`

```python
available_agents['os_user'] = np.where(available_agents.a.str.contains('Windows', flags= re.IGNORECASE),
                              'Windows', 'Non-Windows')
```

```python
pd.crosstab(available_agents.os_user, available_agents.tz).T.sort_values(by= 'Windows', ascending= False)
```

or

```python
available_agents.groupby(['tz', 'os_user']).size().unstack().fillna(0).sort_values(by= 'Windows', ascending= False)
```

# 

```python
tz_os = available_agents.groupby(['tz', 'os_user']).size().unstack().fillna(0)
tz_os
```

```python
top_indices = tz_os.sum(1).argsort()
top_indices
```

#### What is argsort?'

# 

```python
largest_10 = tz_os.take(top_indices[-10:])
largest_10
```

```python
largest_10 = largest_10.stack()
largest_10.name = 'total'
largest_10 = largest_10.reset_index()
```

```python
largest_10
```

```python
sns.barplot(y= 'tz', x= 'total', data= largest_10, hue= 'os_user')
```

As this is not NORMALIZED, we will normalize and make them out of 100%.

### Way 1 with Transform 

```python
def normalize(group):
    return group / group.sum()
```

```python
largest_10['outof1'] = largest_10.groupby(['tz']).transform(normalize)
largest_10
```

```python
sns.barplot(y= 'tz', x= 'outof1', data= largest_10, hue= 'os_user')
```

### Way 2 with Apply 
Remember GLUE part.

```python
def normalize(group):
    # GLUE part â†“
    group['outout'] = group.total / group.total.sum()
    return group
```

```python
largest_10.groupby('tz').apply(normalize)
```
